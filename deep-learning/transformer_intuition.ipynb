# Trannsformer Intuition Summary

### What problem does self-attention solve?

A normal RNN or LSTM processes data sequentially, meaning it uses previous inputs to help determine the next output. However, information has to pass through many time steps before it reaches later parts of the sequence. As sequences become longer, it becomes increasingly difficult for the model to preserve important information from earlier inputs. This means the model may forget context that is needed to make accurate predictions later in the sequence, which can reduce its overall performance on long sequences.

### How does self-attention work?

Self-attention solves this problem by allowing every input to compare itself with every other input in the sequence instead of processing them one at a time. This gives each input access to all the relevant information in the sequence, rather than relying only on information passed through previous time steps. The model determines which inputs are most important by computing attention weights that measure how relevant each input is to the current prediction. Inputs with higher attention weights contribute more to the final representation, allowing the model to focus on the most useful information while still considering the entire sequence.

### What is a transformer?

A Transformer is a deep learning architecture that uses self-attention as its main mechanism for processing sequential information. Unlike RNNs and LSTMs, which process inputs one at a time and rely on a hidden state to store previous information, Transformers allow every input in a sequence to directly interact with every other input. By using self-attention, the model can determine which parts of the sequence are most important and create more meaningful representations of the data. Transformers also use additional components such as positional encoding and feedforward neural networks to understand the order and transform the information into more useful forms. This architecture allows models to capture long-range relationships more effectively and has become the foundation for many modern AI systems.
