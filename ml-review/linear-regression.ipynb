# Linear Regression with NumPy


## 1. Import Libraries

NumPy is used for numerical computation and array operations, while Matplotlib is used to visualize the dataset, the learned regression line, and the loss curve during training.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 2. Generate a Synthetic Dataset

To demonstrate linear regression, a synthetic dataset is generated with a known linear relationship. Random Gaussian noise is added to simulate the variability found in real-world data.

```python
np.random.seed(42)

x = np.linspace(0, 10, 100)
y = 3 * x + 5 + np.random.normal(0, 2, size=len(x))
```

## 3. Visualize the Dataset

Before training a model, it is useful to inspect the data. Plotting the observations confirms the approximately linear relationship between the input and output variables.

```python
plt.figure(figsize=(8,5))
plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Synthetic Dataset")
plt.show()
```

## 4. Initialize Model Parameters

The linear regression model is defined by two parameters:

- **w** — the slope of the line
- **b** — the y-intercept

These parameters are initialized to zero and will be updated during training using gradient descent.

The learning rate controls the size of each update, while the number of epochs determines how many optimization steps are performed.

```python
w = 0
b = 0

learning_rate = 0.01
epochs = 1000

loss_history = []
n = len(x)
```

## 5. Implement Gradient Descent

Gradient descent is an optimization algorithm that minimizes the loss function by repeatedly updating the model parameters.

Each iteration consists of four steps:

1. Compute predictions.
2. Calculate the Mean Squared Error (MSE).
3. Compute the gradients with respect to the parameters.
4. Update the parameters in the direction that reduces the loss.

The loss from each iteration is stored so the learning process can be visualized later.

```python
for epoch in range(epochs):
    ## predict values
    y_pred = w * x + b

    ## mean squared error and tracking loss
    loss = (1/n) * np.sum((y - y_pred) ** 2)
    loss_history.append(loss)

    ## calculate gradients
    dw = -(2/n) * np.sum(x * (y - y_pred))
    db = -(2/n) * np.sum(y - y_pred)

    ## apply gradients to slope and y-intercept
    w -= learning_rate * dw
    b -= learning_rate * db
```

## 6. Evaluate the Learned Model

After training, the learned parameters can be examined. Ideally, the model should recover values close to the parameters used to generate the dataset.

The final Mean Squared Error provides a quantitative measure of the model's performance.

```python
print(f"Slope (w): {w:.3f}, Intercept (b): {b:.3f}")
print(f"Final Loss: {loss_history[-1]:.3f}")
```

## 7. Plot the Learned Regression Line

The learned regression line is plotted alongside the original data points. If training was successful, the line should closely match the underlying trend in the data.

```python
plt.figure(figsize=(8,5))
plt.scatter(x, y, label="Data")
plt.plot(x, w * x + b, color="red", linewidth=2, label="Regression Line")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression Fit")
plt.legend()

plt.show()
```

## 8. Visualize the Training Process

The loss curve illustrates how the Mean Squared Error changes throughout training. A steadily decreasing loss indicates that gradient descent is successfully optimizing the model parameters.

```python
plt.figure(figsize=(8,5))
plt.plot(loss_history)

plt.xlabel("Epoch")
plt.ylabel("Mean Squared Error")
plt.title("Gradient Descent Loss Curve")

plt.show()
```
