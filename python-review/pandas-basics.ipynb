# Pandas Fundamentals

This notebook reviews the core features of Pandas for working with structured data in Python. Pandas provides powerful tools for loading, cleaning, filtering, transforming, and analyzing tabular datasets through the DataFrame data structure. The concepts covered in this notebook—including DataFrames, filtering, grouping, and merging—form the foundation of exploratory data analysis and data preprocessing used throughout machine learning workflows. To reinforce these concepts, this notebook explores the Titanic dataset, one of the most widely used datasets for learning data analysis techniques.

## DataFrames

The primary data structure in Pandas is the **DataFrame**, a two-dimensional table consisting of rows and columns. Unlike NumPy arrays, DataFrames can store multiple data types and provide labeled rows and columns, making them ideal for working with real-world datasets.

DataFrames support efficient data manipulation, statistical analysis, and integration with other scientific computing libraries. Most machine learning datasets are initially loaded into a DataFrame before being cleaned and transformed for modeling.

```python
import pandas as pd

# Create a dataframe from a dictionary
students = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [16, 17, 16],
    "Grade": [92, 85, 98]
}

df = pd.DataFrame(students)

print(df)

print(df.shape)
print(df.columns)
print(df.dtypes)
```

## Filtering

Filtering allows subsets of a dataset to be selected based on logical conditions. Rows can be filtered using comparison operators, multiple conditions, and boolean expressions. Filtering is one of the most common preprocessing operations and is frequently used to remove missing values, isolate specific groups, or create training subsets.

```python
students = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Grade": [92, 85, 98, 78],
    "Age": [16, 17, 16, 18]
}

df = pd.DataFrame(students)

print(df[df["Grade"] > 90])

print(f"\n{df[df["Age"] > 16]}")

print(f"\n{df[(df["Grade"] > 95) & (df["Age"] == 16)]}")
```

## GroupBy

The `groupby()` operation partitions a dataset into groups based on one or more columns and applies aggregate functions to each group. This enables efficient calculation of summary statistics such as averages, counts, sums, and medians. Grouping data is a fundamental technique in exploratory data analysis for identifying trends and relationships within a dataset.

```python
students = {
    "Class": ["A", "A", "B", "B", "B"],
    "Grade": [90, 85, 88, 95, 80]
}

df = pd.DataFrame(students)

print(df.groupby("Class")["Grade"].mean())


print(f"\n{df.groupby("Class").size()}")

print(f"\n{df.groupby("Class")["Grade"].agg(["mean", "max", "min"])}")
```

## Merge

Datasets are often stored across multiple tables that must be combined before analysis. The `merge()` function joins DataFrames based on one or more shared columns, similar to SQL joins. Understanding how to merge datasets is essential for integrating information from multiple sources during data preprocessing.

```python
students = pd.DataFrame({
    "StudentID": [101, 102, 103],
    "Name": ["Alice", "Bob", "Charlie"]
})

grades = pd.DataFrame({
    "StudentID": [101, 102, 103],
    "Grade": [92, 85, 98]
})

# Merge on StudentID
merged = pd.merge(students, grades, on="StudentID")

print(merged)
```

## Titanic Data Analysis

```python
import pandas as pd
import numpy as np

df = pd.read_csv("titanic.csv")

print("First 5 Rows")
print(df.head())

print("\nDataset Shape")
print(df.shape)

print("\nColumn Names")
print(df.columns)

print("\nData Types")
print(df.dtypes)

print("\nDataset Information")
print(df.info())

print("\nSummary Statistics")
print(df.describe())


print("\nMissing Values")
print(df.isnull().sum())

# Survival statistics
overall_survival = df["Survived"].mean()

print(f"\nOverall Survival Rate: {overall_survival:.2%}")

print("\nSurvival Rate by Passenger Class")

print(
    df.groupby("Pclass")["Survived"].mean()
)

print("\nSurvival Rate by Sex")

print(
    df.groupby("Sex")["Survived"].mean()
)

# Passenger class data
print("\nAverage Fare by Passenger Class")

print(
    df.groupby("Pclass")["Fare"].mean()
)


print("\nAverage Age by Passenger Class")

print(
    df.groupby("Pclass")["Age"].mean()
)

# Filtering
print("\nFirst Class Passengers")

print(
    df[df["Pclass"] == 1].head()
)

print("\nPassengers Under 18")

print(
    df[df["Age"] < 18].head()
)

print("\nFemale Survivors")

print(
    df[
        (df["Sex"] == "female") &
        (df["Survived"] == 1)
    ].head()
)


df["IsChild"] = df["Age"] < 18

print("\nChild Survival Rate")

print(
    df.groupby("IsChild")["Survived"].mean()
)


print("\nTop 10 Highest Fares")

print(
    df.sort_values(
        by="Fare",
        ascending=False
    )[["Name","Fare"]].head(10)
)

# Merge example
extra = pd.DataFrame({

    "PassengerId": df["PassengerId"],

    "BoardingPriority": np.random.choice(
        ["High","Medium","Low"],
        size=len(df)
    )

})

merged = pd.merge(
    df,
    extra,
    on="PassengerId"
)

print("\nMerged Dataset")

print(
    merged.head()
)

print("\n========== SUMMARY ==========")

print(f"Total Passengers: {len(df)}")

print(f"Overall Survival Rate: {overall_survival:.2%}")

print(f"Average Age: {df['Age'].mean():.2f}")

print(f"Average Fare: ${df['Fare'].mean():.2f}")

print(f"Highest Fare: ${df['Fare'].max():.2f}")

print(f"Lowest Fare: ${df['Fare'].min():.2f}")
```
