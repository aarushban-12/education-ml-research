### Problem 1
Given a list of integers, return a new list containing only the even numbers using a list comprehension.

```python
nums = [1,2,3,4,5,6,2,2,1,4,1]

even_nums = [num for num in nums if num % 2 == 0]
print(even_nums)
```

### Problem 2
Write a function that returns a dictionary where each key is a word and each value is the number of times it appears, ignore punctuation.

```python
text = "hello one two hello three one four"

def count_words(text):
    words = text.split()
    word_count = {}
    for word in words:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1
    return word_count

print(count_words(text))
```

### Problem 3
Without using sort() or sorted() find the second largest number in a list.

```python
nums = [1,4,2,4,2,12,42,41,63,19,30]

maxi = max(nums)
mini = min(nums)

small_diff = maxi - mini
val = nums.index(mini)

for index, item in enumerate(nums):
    if maxi - item < small_diff and maxi-item != 0:
        small_diff = maxi - item
        val = index

print(nums[val])
```

### Problem 4
Given a dictionary, return a dictionary mapping each student to their average score.


```python
grades = {
    "Alice": [92, 88, 95],
    "Bob": [81, 79, 85],
    "Charlie": [100, 98, 96]
}

avg_dict = {key: round(sum(value)/len(value), 2) for key, value in grades.items()}

print(avg_dict)
```

### Problem 6
Write a function that takes a two-dimensional list (matrix) of numbers and returns a list containing the sum of each row. Then modify your solution to also return the average of each row.

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

sum_rows = [sum(row) for row in matrix]
print(sum_rows, sum(sum_rows)/len(sum_rows))
```

### Problem 7
Write a function that finds and returns the element that appears most frequently in a list. If multiple elements are tied for the highest frequency, return any one of them.

```python
nums = [1,2,3,2,1,2,3,6,5,3,5,3,22,2,2]

most_frequent = max(set(nums), key=nums.count)

print(most_frequent)
```

### Problem 8
Write a function that takes a list of lists and returns a single flattened list containing all elements in their original order.

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

flattened = [num for row in matrix for num in row]
print(flattened)
```

### Problem 9
Write a function that takes a list of unique integers and returns all integers that are missing between the smallest and largest values in the list.

```python
nums = [1,2,4,7,9,10,12,15,18]

missing = []

for i in range(nums[0], nums[-1]):
    if i not in nums:
        missing.append(i)
print(missing)       
```

### Problem 10
Given a list of tuples representing student interactions, where each tuple contains a student ID and whether their answer was correct, write a function that computes each student's overall accuracy and returns the results in a dictionary.

```python
attempts = [
    (101,1),
    (101,0),
    (102,1),
    (103,1),
    (102,0),
    (101,1),
    (103,0),
    (102,1)
]

percent_success = {}

for student_id, success in attempts:
    if student_id not in percent_success:
        percent_success[student_id] = {'success': 0, 'total': 0}
    percent_success[student_id]['total'] += 1
    if success == 1:
        percent_success[student_id]['success'] += 1

scores = {student_id: round(data['success']/data['total']*100, 2) for student_id, data in percent_success.items()}   
print(scores)     
```
