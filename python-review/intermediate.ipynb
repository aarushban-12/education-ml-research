## Classes

Classes are a fundamental feature of object-oriented programming (OOP) that allow related data and behavior to be grouped into a single object. A class acts as a blueprint for creating objects, each with its own attributes (data) and methods (functions).

Classes improve code organization, reusability, and readability, making them especially useful for larger software projects. In machine learning libraries such as PyTorch, models, datasets, and optimizers are all implemented as classes.

### Key Concepts

- **Class** – A blueprint for creating objects.
- **Object** – An instance of a class.
- **Attributes** – Variables that store an object's state.
- **Methods** – Functions defined inside a class.
- **Constructor (`__init__`)** – Initializes an object when it is created.
- **Inheritance** – Allows one class to inherit functionality from another.
- **Encapsulation** – Groups related data and methods together.

```python
class Student:
    def __init__(self, student_id, name):
        self.student_id = student_id
        self.name = name
        self._score = 0

    def update_score(self, points):
        """Increase the student's score."""
        self._score += points

    def get_score(self):
        """Safely access the student's score."""
        return self._score

class HonorsStudent(Student):
    def __init__(self, student_id, name, school):
        super().__init__(student_id, name)
        self.school = school

    def display_info(self):
        print(f"Name: {self.name}")
        print(f"School: {self.school}")
        print(f"Score: {self.get_score()}")

student = HonorsStudent(101, "Alice", "North High")
student.update_score(15)
student.display_info()
```

## Decorators

Decorators provide a way to modify or extend the behavior of a function without changing its source code. They are higher-order functions that take another function as input and return a modified version of that function.

Decorators are commonly used for logging, timing code execution, caching results, authentication, and measuring model performance. Many Python libraries use decorators extensively.

### Key Concepts

- Functions are first-class objects and can be passed as arguments.
- A decorator wraps another function.
- The `@decorator_name` syntax applies a decorator.
- `functools.wraps` preserves the original function's metadata.

```python
from functools import wraps
import time

def logger(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        print(f"Calling function: {func.__name__}")
        start = time.time()
        time.sleep(1)
        result = func(*args, **kwargs)
        elapsed = time.time() - start
        print(f"Function {func.__name__} completed.")
        print(f"Execution: {elapsed:.4f} seconds")
        return result
    return wrapper

@logger
def greet():
    print("Hello!")

greet()
```

## Generators

Generators are functions that produce values one at a time using the `yield` keyword instead of returning all values at once. Because values are generated only when needed, generators are memory-efficient and well suited for processing large datasets.

Machine learning frameworks frequently use generators to load data in batches instead of loading an entire dataset into memory.

### Key Concepts

- `yield` pauses a function while preserving its state.
- Generator objects are iterable.
- Values are produced lazily (on demand).
- Generators are more memory-efficient than storing every value in a list.

```python
def countdown(n):
    while n > 0:
        yield n
        n-=1

for i in countdown(5):
    print(i)        
```

## File Input and Output (File I/O)

File I/O allows programs to read data from files and write data back to disk. This is essential for loading datasets, saving trained models, recording experiment results, and storing configuration files.

Python provides built-in support for working with text files, CSV files, and JSON files.

### Key Concepts

- Open files using `with open(...)` to ensure they are closed automatically.
- Read files using methods such as `read()`, `readline()`, and `readlines()`.
- Write files using `write()` or `writelines()`.
- JSON is commonly used for structured data.
- CSV files are commonly used for tabular datasets.

```python
import json

data = {
    "name": "Alice",
    "score": 95
}

with open("student.json", "w") as file:
    json.dump(data, file)

with open("student.json", "r") as file:
    loaded = json.load(file)

print(loaded)
```
